```python
#importing draw_ocr for annotations and PaddleOCR class
from paddleocr import PaddleOCR, draw_ocr

#for plotting images
import matplotlib.pyplot as plt

#opencv
import cv2

#folder directory navigation
import os
```

```python
import paddleocr
paddleocr.__version__
```

```python
#setup model
ocr_model = PaddleOCR(lang='en')
```

```python
#providing path of data
img_path = os.path.join('./data', 'nimsulide.jpeg')
```

```python
img_path
```

```python
# performing OCR
result = ocr_model.ocr(img_path)
```

```python
#printing all the box coordinates, extracted text and accuracy
#the result is printed in this format: list[list[list(x,y) list(x,y) list(x,y) list(x,y)], tuple('text', accuracy) ]]
result
```

```python
result[0][0][1][0] #first extracted text
```

```python
#printing all the text
for res in result:
  for re in res:
    print(re[1][0])
```

```python
#storing all the result in  a variable
box = [re[0] for re in res]
text = [re[1][0] for re in res]
score = [re[1][1] for re in res]
```

```python
score
```

```python
#providing path of the fonts
fonts = os.path.join('PaddleOCR','doc','fonts','nepali.ttf')
```

```python
#imports image
img = cv2.imread(img_path)

#reorders color channels
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
#resizing display area
plt.figure(figsize=(55, 55))

#draw annotations on image
annotated = draw_ocr(img, box, text, score, font_path=fonts)

#show image using matplotlib
plt.imshow(annotated)
```

```python
#storing all the extracted text in  a line
output = " "
for res in result:
  for re in res:
    output += re[1][0] + " "

print(output)
```

```python
# Create a text file and write the content
with open('outputs.txt', 'w') as f:
    f.write(output)

# Download the text file
from google.colab import files
files.download('outputs.txt')
```
