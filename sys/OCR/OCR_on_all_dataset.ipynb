```python

import os

# Set the path to the directory containing your images
image_dir = 'dataset'

# Get a list of all image files in the directory
image_files = os.listdir(image_dir)

# Sort the image files
image_files.sort()

# Set the starting number for renaming
start_number = 1

# Rename the images
for i, old_name in enumerate(image_files):
    # Create the new name by adding the start number and maintaining the file extension
    new_name = f"{start_number + i:04d}{os.path.splitext(old_name)[1]}"

    # Construct the full paths for the old and new names
    old_path = os.path.join(image_dir, old_name)
    new_path = os.path.join(image_dir, new_name)

    # Rename the file
    os.rename(old_path, new_path)

    print(f"Renamed: {old_name} -> {new_name}")
```

```python
#importing
from paddleocr import PaddleOCR, draw_ocr
import matplotlib.pyplot as plt
import PIL
import cv2
```

```python
#setup model
ocr_model = PaddleOCR(lang='en')
```

```python
img_path = 'dataset'
img_files = os.listdir(img_path)
sorted_img_files = sorted(img_files)
```

```python
ocr_results = []
file_name =[]
# Get a sorted list of image paths
sorted_img_paths = sorted(os.listdir(img_path))

# Iterate over each sorted image path
for img_name in sorted_img_paths:
    # Construct the full path to the current image
    img = os.path.join(img_path, img_name)

    # Perform OCR on the current image and store the result
    result = ocr_model.ocr(img)

    # Append the OCR result to the list
    ocr_results.append(result)
    file_name.append(img_name)
```

```python
#storing all the text, box and score in list
text_list = []
box_list = []
score_list = []
for resu in ocr_results:
    for res in resu:
        for re in res:
            text_value = re[1][0]
            text_list.append(text_value)

            box_value = re[0]
            box_list.append(box_value)

            score_value = re[1][1]
            score_list.append(score_value)

print(text_list)
print(box_list)
print(score_list)
```

```python
#storing all the extracted text in  a line for annotation
output = " "
for resu in ocr_results:
  for res in resu:
    for re in res:
      output += re[1][0] + " "
  output = output + "\n\n"

print(output)
```

```python
output = ""
paragraph_number = 1

for resu in ocr_results:
    for res in resu:
        # Print the paragraph number at the beginning of each paragraph
        output += f"Image {paragraph_number}: "

        for re in res:
            output += re[1][0] + " "

        output += "\n\n"
        paragraph_number += 1

print(output)
```

```python
# Specify the file path where you want to save the output
file_path = "output.txt"

# Write the output to the file
with open(file_path, 'w') as file:
    file.write(output)

# Print a message indicating that the file has been saved
print(f"Output has been saved to {file_path}")
```

```python
import os
from zipfile import ZipFile

# Specify the folder path containing the images
folder_path = 'dataset'

# Create a list of image file names in the folder
image_files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f)) and f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp'))]

# Create a zip file to store the images
zip_file_path = 'images.zip'
with ZipFile(zip_file_path, 'w') as zipf:
    for image_file in image_files:
        image_path = os.path.join(folder_path, image_file)
        zipf.write(image_path, arcname=image_file)

# Print a message indicating that the zip file has been created
print(f"All images have been zipped into {zip_file_path}")
```
